```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-100, 100, 100)
y = (np.exp(np.sqrt(2*t)) - np.exp(-np.sqrt(2*t))) * 1/(np.sinh(2*np.sqrt(2)))

plt.plot(t,np.imag(y),color='blue',label="imaginery",linewidth=1.5)
plt.plot(t,np.real(y),color='yellow',label='real',linewidth=1.5)
plt.plot(t,y,label='whole function',color='red',linewidth=2)

plt.xlabel('t')
plt.ylabel('y')
plt.title('Plot of y and its Real and Imaginary parts')
plt.legend(['Imaginary(y)', 'Real(y)', 'y'])
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

t = np.linspace(-100, 100, 100)
y = (np.exp(np.sqrt(2*t)) - np.exp(-np.sqrt(2*t))) * 1/(np.sinh(2*np.sqrt(2)))

T, Y = np.meshgrid(t, np.imag(y))
Z = np.zeros_like(T)

fig = plt.figure()
ax = plt.axes(projection='3d')

# Plot the imaginary part as a surface
ax.plot_surface(T, Y, Z, cmap='Blues', alpha=0.8)

# Plot the real part as a surface
R, Y = np.meshgrid(t, np.real(y))
ax.plot_surface(R, Y, Z, cmap='copper', alpha=0.8)

# Plot the whole function as a line
ax.plot(t, np.imag(y), y, color='red', linewidth=2)

ax.set_xlabel('t')
ax.set_ylabel('Imaginary(y) / Real(y)')
ax.set_zlabel('y')
ax.set_title('3D Plot of y and its Real and Imaginary parts')

plt.show()


```
